```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
df.dropna(axis = 1)    #deletes all the columns with missing values as "deck and embark_town"
df.dropna(axis = 0)    #deletes all the rows with missing values
```

## Mean Imputation - replacing missing values with mean of the column
### 1-Mean Value

```python
df['age'].fillna(df['age'].mean())    #fills the missing values with the mean of the column
```

```python
df2 = sns.load_dataset('titanic')
```

```python
df2['age'].fillna(df2['age'].mean(),inplace= True)
df2.isnull().sum()
```

```python
df['age_new'] = df['age'].fillna(df['age'].mean())
```

```python
df.head()
```

```python
df[['age_new','age']].isnull().sum()
```

## Median when we have outliers

```python
df['age_median'] = df['age'].fillna(df['age'].median())
df
```

```python
df['age_median'].isnull()
```

## Mode Imputation technique

```python
df['embarked'].isnull()
```

```python
df[df['embarked'].isnull()]
```

```python
df.columns
```

```python
df['embarked'].unique()
```

```python
mode_val = df[df['embarked'].notna()]['embarked'].mode()[0]  #mode() returns a series, so we need to get the first value
```

```python
df['embarked_new'] = df['embarked'].fillna(mode_val) #filling the missing values with the mode value
```

```python
df[['embarked_new' , 'embarked']].isnull().sum() #checking if the missing values are filled
```
