```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip',data=tips)
plt.title('Scatter plot of total bill and tip')
```

```python
sns.lineplot(x='size',y='total_bill' ,data=tips)
```

```python
#category plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill and day')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.title('Sex to total bill')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'] ,  bins = 10 , kde = True)   #kde = True is for density plot
plt.title('Histogram of total bill')                         #kde stands for kernel density estimation
```

```python
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
#pair plot
sns.pairplot(tips)
```

```python
corr = tips[['total_bill' , 'tip' , 'size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm') #annot is for showing the values in the heatmap
plt.title('Heatmap of correlation matrix')  #cmap is for color map 
```
