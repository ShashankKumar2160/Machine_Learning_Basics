```python
### MERGING AND JOING DFS
```

```python
df1 = pd.DataFrame({
    'key1' : ['A','B','C'],
    'value1' : [1,2,3]
})
df2 = pd.DataFrame({
    'key1' : ['A','B','D'],
    'value1' : [4,5,6]
})

print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2,on='key1',how='inner')) # inner join to get common values
print(pd.merge(df1,df2,on='key1',how='outer')) # outer join to get all values
```

```python
print(pd.merge(df1,df2,on='key1',how='left')) # left join to get all values of left df
print(pd.merge(df1,df2,on='key1',how='right'))# right join to get all values of right df
```
